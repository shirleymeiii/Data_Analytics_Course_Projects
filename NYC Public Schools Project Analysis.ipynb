```python
#importing all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#reading the dataset
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')

#viewing the first 10 rows
df.head(10)
```

```python
#identify the number of rows and columns
df.shape
```

```python
#column names
df.columns
```

```python
#adding new column for borough which is represented by the following letters indicated below 
#M is Manhattan, K is Brooklyn, Q is Queens, R is Staten Island, and X is Bronx
df['Borough']=df['School DBN'].str[2]
#viewing data to check newly added column
df.head()
```

```python
#view datatypes
df.dtypes
```

```python
#converting columns from object to numeric
convertcolumns=['Mean Score','Number Scoring Below 65','Percent Scoring Below 65','Number Scoring 65 or Above','Percent Scoring 65 or Above','Number Scoring 80 or Above','Percent Scoring 80 or Above']
df[convertcolumns]=df[convertcolumns].apply(pd.to_numeric,errors='coerce')
```

```python
#viewing datatypes again
df.dtypes
```

```python
#viewing df
df
```

```python
#replace missing values
missing_values=['s','na','--']
df.replace(missing_values,float('NaN'),inplace=True)
#viewing df
df
```

```python
#checking if there any null values 
df.isnull().sum()
```

```python
#find the rows with null values under 'Regent Exam' column
df[df['Regents Exam'].isnull()]
```

```python
#Removing all Null values in df_new
df_new = df.dropna(axis=0)
```

```python
#checking for null values in df_new 
df_new.isnull().sum()
```

```python
#viewing df_new
df_new
```

```python
#finding the mean scores by school level
mean_scores=df_new.groupby(['Year','Borough'])['Mean Score'].mean()
mean_scores
```

```python
#filtering by high schools in Brooklyn only
brooklyn_highschools = df_new[(df_new['Borough']=='K') & (df_new['School Level']=='High school')]
brooklyn_highschools
```

```python
#Using Midwood Highschool as a comparison point for Common Core Algebra
Midwood_data_algebra=brooklyn_highschools[(brooklyn_highschools['School Name']=='Midwood High School') & (brooklyn_highschools['Regents Exam']=='Common Core Algebra')]
Midwood_data_algebra
```

```python
#Midwood Mean Score by Year for Common Core Algebra
Midwood_algebra_score = Midwood_data_algebra.groupby('Year')['Mean Score'].mean()
Midwood_algebra_score
```

```python
#Filter data for all other brooklyn highschools for common core algebra
otherbk_data_algebra=brooklyn_highschools[(brooklyn_highschools['School Name'] !='Midwood High School') & (brooklyn_highschools['Regents Exam']=='Common Core Algebra')]
otherbk_data_algebra
```

```python
#Other brooklyn highschools mean score for common core algebra
otherbk_algebra_score = otherbk_data_algebra.groupby('Year')['Mean Score'].mean()
otherbk_algebra_score
```

```python
#Combining the dataframes for plotting purposes
Combined_data=pd.concat([Midwood_algebra_score, otherbk_algebra_score], keys=['Midwood High School', 'Other Broolyn High Schools'])

#Plotting comparison graph between Midwood High School and Other Schools

Combined_data.plot(x="Year", kind='bar', color=['slategrey', 'lightsteelblue', 'cornflowerblue'])
plt.title('Midwood High School vs. Other Brooklyn High Schools for Common Core Algebra')
plt.xlabel('Year',fontsize=10)
plt.ylabel('Mean Algebra Score',fontsize=10)
plt.show()

```
